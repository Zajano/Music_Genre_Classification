### loading audio files

```python
song_paths, filenames = make_file_list(music_dir)
```

```python
test_song, test_sr = librosa.load(song_paths[0])
```

### MFCCs

```python
test_gram = librosa.feature.melspectrogram(test_song, sr=test_sr, power=2.0)
test_db = librosa.power_to_db(test_gram)
```

```python
plt.figure(figsize=(15, 5))
librosa.display.specshow(test_db, sr=test_sr, x_axis='time', y_axis='mel')
plt.colorbar()
```
